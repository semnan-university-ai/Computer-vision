Subplot2grid() Function
</h1>
<p style="font-size: 150%;text-align: justify;line-height: 140%;">
This function gives more flexibility in creating an axes object at a specific location of the grid. It also allows the axes object to be spanned across multiple rows or columns.
</p>

<code style="font-size: 150%;text-align: justify;line-height: 140%;">Plt.subplot2grid(shape, location, rowspan, colspan)

<br>



```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)


x = np.arange(1,10)
a2.plot(x, x*x)
a2.set_title('square')
a1.plot(x, np.exp(x))
a1.set_title('exp')
a3.plot(x, np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

<h1 style="font-size: 250%;text-align: justify;line-height: 140%;">
Logarithmic and other nonlinear axes
</h1>
<p style="font-size: 150%;text-align: justify;line-height: 140%;">
matplotlib.pyplot supports not only linear axis scales, but also logarithmic and logit scales. This is commonly used if data spans many orders of magnitude. Changing the scale of an axis is easy:
</p>
<code style="font-size: 150%;text-align: justify;line-height: 140%;">plt.xscale('log')
</code>

<br>



```python

from matplotlib.pyplot import figure

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
figure(num=None, figsize=(10, 10))

# linear
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
```

```python
# log
figure(num=None, figsize=(10, 10))
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
plt.show()
```

```python
# symmetric log
figure(num=None, figsize=(10, 10))

plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)
plt.show()
```

```python
# logit
figure(num=None, figsize=(10, 30))
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
plt.show()
```

<h1 style="font-size: 250%;text-align: justify;line-height: 140%;">
3D Surface plot
</h1>
<p style="font-size: 150%;text-align: justify;line-height: 140%;">
Surface plot shows a functional relationship between a designated dependent variable (Y), and two independent variables (X and Z). The plot is a companion plot to the contour plot. A surface plot is like a wireframe plot, but each face of the wireframe is a filled polygon. This can aid perception of the topology of the surface being visualized. The plot_surface() function x,y and z as arguments.
</p>


<br>



```python
from mpl_toolkits import mplot3d

x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = x.copy().T 
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()

```

```python



from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

x, y = np.mgrid[-1.0:1.0:30j, -1.0:1.0:30j]

xy = np.column_stack([x.flat, y.flat])

mu = np.array([0.0, 0.0])

sigma = np.array([.5, .5])
covariance = np.diag(sigma**2)

z = multivariate_normal.pdf(xy, mean=mu, cov=covariance)

z = z.reshape(x.shape)

fig = plt.figure(num=None, figsize=(10, 10))

ax = fig.add_subplot(111, projection='3d')



ax.plot_surface(x,y,z)

plt.show()
```

Sometimes you want to enhance the contrast in your image, or expand the contrast in a particular region while sacrificing the detail in colors that don't vary much, or don't matter.
```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')


fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,
def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)


anim.save('sine_wave.gif', writer='imagemagick')
```
